```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
import plotly.express as px
```

```python
btc_df = yf.download('BTC-USD', 
                      start='2020-01-01', 
                      end='2021-07-22', 
                      progress=False,
)

egld_df = yf.download('EGLD-USD', 
                      start='2020-01-01', 
                      end='2021-07-23', 
                      progress=False,
)
```

```python
egld_df
```

```python
#fig = go.Figure()
#fig.add_trace(go.Scatter(x=btc_df.index, 
#                         y=btc_df['Close'],
#                         name = 'BTC Price'
#                        ))

fig = px.area(egld_df, y='Close', title="EGLD Close Price")
fig.show()
```

```python
candlestick = go.Figure(data=[go.Candlestick(x=egld_df.index,
                                     open=egld_df['Open'],
                                     high=egld_df['High'],
                                     low=egld_df['Low'],
                                     close=egld_df['Close'],
                                    )])

candlestick.update_layout(title = 'EGLD Candlestick Chart')
candlestick.show()
```
